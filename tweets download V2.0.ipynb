```python
import snscrape.modules.twitter as twitterScraper
import snscrape
import pandas as pd
import pandas
import json
import datetime
from date# Tweets for Ethereum NFTtime import date
```

# Tweets for Ethereum NFT

```python
search_query = "opensea nft"
from_date = "2020-01-01"     #"2018-02-04"
until_date = "2021-04-27"
tweets_opensea_nft = [] #196,471 tweets in total
```

```python
for i,tweet in enumerate(twitterScraper.TwitterSearchScraper("{} since:{} until:{}".format(search_query, from_date, until_date)).get_items()):
  if i>1000000:
    break
  tweets_opensea_nft.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.replyCount, tweet.likeCount, tweet.retweetCount])
```

```python
tweets_opensea_nft
```

```python
df = pd.DataFrame([tweets_opensea_nft])
```

```python
df = df.T
```

```python
df = df.rename(columns={0: "Time"})
```

```python
df
```

```python
df = df.Time.apply(pd.Series)
```

```python
df
```

```python
df = df.rename(columns={0: "time", 1: "id", 2: "body", 3: "username", 4: "replycount", 5: "likecount", 6: "retweetcount"})
```

```python
df
```

```python
df = df.to_csv("opensea_nft_1M.csv", index=True)
```
